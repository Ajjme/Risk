```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom, nbinom, poisson
from scipy.optimize import minimize
```

```python
# Load data from Excel file
data = pd.read_excel('Data.xlsx')  # Reads data from an Excel file

# Compute the total number of observations (sum of observed data)
n = data['Observed Data'].sum()

# Compute relative frequency of observed data and store it in a new column
data['Relative Frequency'] = data['Observed Data'] / n
```

```python
# Define the Poisson distribution parameter (lambda)
poisson_lambda = 2

# Compute the expected counts for each bin using the Poisson distribution
# n * Poisson PMF gives the expected count for each number of made shots
data['Poisson'] = n * poisson.pmf(data['Made shots'], poisson_lambda)

# Calculate bin edges for histogram plotting
# Bin edges are calculated based on bin averages and spacing between them
bin_averages = data['Made shots'].values
bin_edges = np.concatenate((
    [bin_averages[0] - (bin_averages[1] - bin_averages[0]) / 2],  # Lower edge
    (bin_averages[:-1] + bin_averages[1:]) / 2,  # Midpoints between averages
    [bin_averages[-1] + (bin_averages[-1] - bin_averages[-2]) / 2]  # Upper edge
))

# Plot the histogram-like bar chart
plt.bar(
    bin_averages,  # Bin centers
    data['Observed Data'],  # Heights of bars (observed counts)
    width=np.diff(bin_edges),  # Widths of bars based on bin edges
    align='center',  # Align bars at bin centers
    edgecolor='black',  # Add black edges to bars
    alpha=0.7,  # Transparency level
    color='red',  # Bar color
    label='Observed Data'  # Label for the legend
)

# Plot the Poisson distribution (expected counts)
plt.plot(
    data['Made shots'],  # X values
    data['Poisson'],  # Y values (expected counts)
    color='black',  # Line color
    label='Poisson Distribution'  # Label for the legend
)

# Add labels, legend, and display the plot
plt.xlabel('Made 3 Point Shots per Game')
plt.ylabel('Counts')
plt.legend()
plt.show()
```

```python

def LogLikelihood(parameters, data, distribution):
    probabilities = distribution.pmf(data['Made shots'], parameters)  # PMF for each bin
    log_probabilities = -np.log(probabilities)  # Negative log probabilities
    return np.sum(data['Observed Data'] * log_probabilities)  # Weighted sum of log probabilities

# Define a function to calculate the chi-squared statistic
# The chi-squared test measures the difference between observed and expected counts
# Lower values indicate a better fit

def ChiSquared(parameters, data, distribution):
    n = data['Observed Data'].sum()  # Total number of observations
    expected_count = n * distribution.pmf(data['Made shots'], parameters)  # Expected counts
    return np.sum((data['Observed Data'] - expected_count) ** 2 / expected_count)  # Chi-squared formula

# Example usage of the ChiSquared function with an initial parameter
parameters = [2]
print(ChiSquared(parameters, data, poisson))
```

```python
# Define a range of lambda values to evaluate the functions
lambda_values = np.linspace(1, 5, 100)
loglikelihood_values = []  # Store negative log-likelihood values
chisquared_values = []  # Store chi-squared values

# Compute the values of negative log-likelihood and chi-squared for each lambda
for lambda_ in lambda_values:
    parameters = [lambda_]
    loglikelihood_values.append(LogLikelihood(parameters, data, poisson))
    chisquared_values.append(ChiSquared(parameters, data, poisson))

# Plot the negative log-likelihood and chi-squared values
plt.plot(lambda_values, loglikelihood_values, color='red', label='Negative Log Likelihood')
plt.plot(lambda_values, chisquared_values, color='orange', label='Chi Squared')
plt.xlabel('Lambda Value')
plt.ylabel('Function Value')
plt.legend()
plt.show()
```

```python
# Use the minimize function to find the optimal parameter
# Minimize the chi-squared function to find the best lambda
initial_guess = [2]  # Initial guess for lambda
result = minimize(ChiSquared, initial_guess, args=(data, poisson))
print("Optimal Lambda:", result.x)
```

```python
# Define the Chi-Squared function for the negative binomial distribution
def ChiSquaredNB(parameters, data):
    r, p = parameters  # Negative binomial parameters: r (successes), p (probability)
    n = data['Observed Data'].sum()  # Total number of observations
    expected_count = n * nbinom.pmf(data['Made shots'], r, p)  # Expected counts using nbinom PMF
    return np.sum((data['Observed Data'] - expected_count) ** 2 / expected_count)  # Chi-squared formula

# Initial guess for the negative binomial parameters
# Example: r = 2 (successes), p = 0.5 (success probability)
initial_guess = [2, 0.5]

# Minimize the Chi-Squared function to find the optimal parameters
result = minimize(ChiSquaredNB, initial_guess, args=(data,))  # Adding bounds to ensure parameters are valid

# Display the result
optimal_r, optimal_p = result.x
print(f"Optimal Parameters:\n r (successes) = {optimal_r}\n p (probability) = {optimal_p}")
```
